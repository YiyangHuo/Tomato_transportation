```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
IO = "../data/ptab_final.xlsx"

sheet = pd.read_excel(io=IO) 
#we remove date and season year here first because technically it is not "numeric data"
# instead, we need try to find any numeric partern in order to convince ourself to use LSTM
numeric_features = [
                  "week_pross_season",  
                  "#loads", 
                  "worm", 
                  "mold", 
                  "green", 
                  "not_tomato", 
                  "color", 
                  "solids", 
                  "ph"
                   ]
categorical_features = ["county", "tomato_usage"]
date_features = ["date", "season_year"]


#All the data column


mag = (5,2)

# plt.scatter(date,rej_rate,c="#7CAE00",alpha=0.5)
```

```python
First, we want to see whether some of the numerical attributes have potential by finding individual linear regression
```

```python
from sklearn.linear_model import LinearRegression
rej_rate = np.array(sheet["rej_rate"]).reshape(-1, 1)
rej_rate = rej_rate/np.linalg.norm(rej_rate)
index = 0
categorical_features:["county", "tomato_usage"]
plt.figure(figsize=(24,56))
for key in numeric_features:
    if key != "rej_rate":
        np_variable = np.array(sheet[key])
        norm = np_variable / np.linalg.norm(np_variable)
        norm = norm.reshape(-1, 1)
        reg = LinearRegression().fit(norm, rej_rate)
        y_pred = reg.predict(norm)
        print("score", reg.score(norm, rej_rate))
        ax = plt.subplot2grid(mag,(index%mag[0],int(index/mag[0])),colspan=1,rowspan=1)
        ax.set_xlabel(key)
        ax.set_ylim([0,0.02])
        plt.scatter(norm,rej_rate,c="#7CAE00",alpha=0.5)
        plt.plot(norm, y_pred, color='blue', linewidth=3)
        index += 1
```

```python
# Next, we are going to find out multiple linear regression.
```

```python
x_multi = np.array(sheet[numeric_features])
x_multi = x_multi / np.linalg.norm(x_multi)
multi_linear = LinearRegression()
multi_linear.fit(x_multi, rej_rate)
multi_predict = multi_linear.predict(x_multi)
print("coeff", multi_linear.coef_)
print("score", multi_linear.score(x_multi, rej_rate, sample_weight=None))
```
